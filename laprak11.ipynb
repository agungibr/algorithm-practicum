###**Import Dataset**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

###**Import Dataset**

```python
df = pd.read_csv("/content/dataset.csv")
```

###**Line Chart**

```python
plt.plot(df['bulan'], df['indonesia'], label='Indonesia')
plt.plot(df['bulan'], df['japan'], label='Japan')

plt.xlabel('Month')
plt.ylabel('Confirmation Covid Cases')
plt.title('Indonesia vs Japan')

plt.xticks(df['bulan'], ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

plt.legend()
plt.show()
```

###**Scatter Plot**

```python
plt.scatter(df['indonesia'], df['japan'], color='red', label='Indonesia')
plt.scatter(df['japan'], df['indonesia'], color='blue', label='Japan')

plt.title('Indonesia vs Japan')

plt.grid(True)
plt.legend()
plt.show()
```

###**Bar Plot**

```python
plt.bar(df['bulan'] - 0.2, df['indonesia'], width=0.4, label='Indonesia', align='center')
plt.bar(df['bulan'] + 0.2, df['japan'], width=0.4, label='Japan', align='center')

plt.xlabel('Month')
plt.ylabel('Confirmation Covid Cases')
plt.title('Indonesia vs Japan')
plt.xticks(df['bulan'], ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

plt.legend()
plt.show()
```

###**Box Plot**

```python
fig, ax = plt.subplots(figsize=(10, 7))

ax.boxplot(df['indonesia'], positions=[1], widths=0.6, patch_artist=True, boxprops=dict(facecolor='skyblue'))
ax.boxplot(df['japan'], positions=[2], widths=0.6, patch_artist=True, boxprops=dict(facecolor='orange'))

ax.set_xticks([1, 2])
ax.set_xticklabels(['Indonesia', 'Japan'])
ax.set_ylabel('Confirmation Covid Cases')
ax.set_title('Indonesia vs Japan')

plt.show()
```

###**Pie Chart**

```python
indo = df['indonesia'].sum()
jap = df['japan'].sum()

labels = ['Indonesia', 'Japan']
sizes = [indo, jap]
colors = ['skyblue', 'orange']

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90)
plt.title('Indonesia vs Japan')

plt.show()
```
