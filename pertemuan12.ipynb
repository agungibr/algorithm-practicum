###**Import Library**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

###**Line Chart**

```python
yields_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
plt.plot(yields_apples)
```

```python
years = [2010, 2011, 2012, 2013, 2014, 2015]
yields_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
plt.plot(years, yields_apples)
```

```python
years = [2010, 2011, 2012, 2013, 2014, 2015]
yields_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
plt.plot(years, yields_apples)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
```

```python
years = range(2000, 2007)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934]
oranges = [0.962, 0.941, 0.93, 0.923,0.918, 0.908, 0.907]

plt.plot(years, apples)
plt.plot(years, oranges)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
```

###**Scatter Plot**

```python
price = [2.50, 1.23, 4.02, 3.25, 5.00, 4.40]
sales_per_day = [34, 62, 49, 22, 13, 19]

plt.scatter(price, sales_per_day)
plt.show()
```

###**Bar Plot**

```python
data = {
    'country': ['A', 'B', 'C', 'D', 'E'],
    'gdp_per_capita': [45000, 42000, 52000, 49000, 47000]
}

df = pd.DataFrame(data)

colors = ['green', 'blue', 'purple', 'brown', 'teal']
plt.bar(df['country'], df['gdp_per_capita'], color=colors)
plt.title('Country vs GDP Per Capita')
plt.xlabel('Country', fontsize=14)
plt.ylabel('GDP Per Capita', fontsize=14)
plt.show()
```

###**Box Plot**

```python
np.random.seed(10)
data = np.random.normal(100, 20,200)

fig = plt.figure(figsize=(10,7))
plt.boxplot(data)
plt.show()
```

###**Histogram**

```python
nilai_siswa = [80,75,60,95,82,60,87,67,87,70,90,87,80,66,70,75,77]
bin_width = 10
kelompok_interval = range(50,100 + bin_width, bin_width)

plt.hist(nilai_siswa, kelompok_interval)
plt.title('Nilai Siswa', fontsize=12)
plt.xlabel('Nilai', fontsize=12)
plt.ylabel('Nilai', fontsize=12)
plt.grid(True)
```

###**Pie Chart**

```python
y = np.array([35,25,25,15])
mylabels = ['Apples', 'Bananas', 'Cherries', 'Kuldi']

plt.pie(y, labels = mylabels)
plt.show()
```
